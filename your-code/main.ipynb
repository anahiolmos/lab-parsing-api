```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url_initial = 'https://rickandmortyapi.com/api/'

response_initial = requests.get(url_initial)
results_initial = response_initial.json()
results_initial
```

```python
url_character = 'https://rickandmortyapi.com/api/character'

response_char = requests.get(url_character)
results_char = response_char.json()
results_char
```

```python
flattened_data = pd.json_normalize(results_char)
flattened_data
```

```python
pd.json_normalize(flattened_data.results[0])
```

```python
# all code

total_pages = results_char['info']['pages']

df = pd.DataFrame()

for i in range(total_pages):

    url =  'https://rickandmortyapi.com/api/character'
    
    response = requests.get(url)
    
    results = response.json()
    
    flattened_data = pd.json_normalize(results)
    
    flattened_data = pd. json_normalize(flattened_data.results[0])
    
    flattened_data = flattened_data.set_index('id')
    
    df = df.append(flattened_data)

df
```
